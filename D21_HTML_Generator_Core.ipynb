```
# This is a reference script for the Gem to use when generating HTML.
# It ensures the CSS and Layout structure is identical every time.

def generate_d21_html(content_data):
    html_template = f"""
    <!DOCTYPE html>
    <html>
    <head>
        <!-- D21 CSS Framework -->
        <style>
            :root {{
                --blue: #00529B; --red: #E31837; --dark: #1F2937;
            }}
            body {{ font-family: 'Inter', sans-serif; background: #f4f6f8; }}
            .slide {{ width: 1280px; height: 720px; background: white; border-top: 8px solid var(--blue); }}
            h1 {{ font-family: 'Montserrat'; color: var(--blue); text-transform: uppercase; }}
            .metric {{ color: var(--red); font-weight: bold; }}
        </style>
    </head>
    <body>
        <!-- Dynamic Content Injected Here -->
        {content_data}
    </body>
    </html>
    """
    return html_template
```
